### Step 1. Making the data messy

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
```

```python
wine = df.drop(df.columns[[0, 3, 5, 8, 11, 12, 13]], axis = 1)
```

```python
wine.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
```

```python
wine.iloc[0:4, 0] = np.nan
```

```python
wine.iloc[3:5, 3] = np.nan
```

```python
wine.iloc[-30:, 6] = np.nan
```

```python
wine['flavanoids'] = wine['flavanoids'].apply(lambda val: str(val))
```

### Step 1. Count the number of missing values

```python
sum(wine.isna().sum())
```

### Step 2.  How many missing values do we have in 'magnesium'?

```python
wine['magnesium'].isna().sum()
```

### Step 3. Print only the non-null values in alcohol

```python
wine['alcohol'].dropna()
```

### Step 4. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine.fillna({'alcohol': 10, 'magnesium': 100}, inplace=True)
wine.head()
```

### Step 5. Drop all Rows where 'hue' is null

```python
wine.dropna(subset=['hue'], inplace=True)
```

### Step 6: Change 'flavanoids' type to float64 rather than string

```python
wine['flavanoids'] = wine['flavanoids'].apply(lambda x: float(x))
wine.head()
```

### Step 7.  Reset the index, so it starts with 0 again

```python
wine.reset_index()
wine
```
