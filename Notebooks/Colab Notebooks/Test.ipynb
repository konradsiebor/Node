**Create function that determines which rows have "AVE" within it**

```
import numpy as np
import pandas as pd
```

```
dogs = pd.read_csv('https://raw.githubusercontent.com/benartuso/nodef19/master/data/dog_tags.csv', header=0)
```

```
dogs.head()
```

```
dogs['Ave?'] = dogs['StreetName'].apply(lambda x: 'Ave' in str(x))
```
