```
import pandas as pd
```

```
#Change header
dogs = pd.read_csv('https://raw.githubusercontent.com/benartuso/nodef19/master/data/01-intro-to-pandas/dog_tags.csv', header = 0)
dogs.head()
```

```
dogs.columns = ['id', 'fee', 'exp_date', 'street', 'number', 'type', 'year', 'block']
```

### Minor Data Cleaning

Some null values exist in the 'street' column so to make this exercise easier we're going to remove them

```
dogs_clean = dogs.dropna(subset=['street'])
```

# Pandas Exercises Solutions

#### 1. How many tags are given for each year?

```
dogs_clean['year'].value_counts()
```

#### 2. How many tags are there per street type? (AVE, RD, etc)

```
street_type = dogs_clean['street'].apply(lambda val: val.split(' ')[-2])
street_type.value_counts()
```

### 3. How much money was spend in 2015 on dog tags? 

```
tags_2015 = dogs_clean[dogs_clean['year'] == 2015]
fees_2015 = tags_2015['fee']
sum(fees_2015)
```
